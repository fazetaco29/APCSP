```python
#Luke Ellis, Individual project, RELATION OF STOCK PRICES AND COVID, I have airlines, hotels, automotive matufatures and more.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas_datareader.data import DataReader
import yfinance as yf #This is the data i'm importing from yahoo Finance, Its a website not a .csv
from datetime import datetime
```

```python
stock_symbol = 'AAL'
start_date = '2020-01-01' #start date
end_date = '2021-01-01' #end date
stock_data = yf.download(stock_symbol, start=start_date, end=end_date) #gets the date if and transforms it to a varibale
plt.figure(figsize=(12, 6)) #gets size of graph
sns.set(style="whitegrid") #makes the grid
plt.plot(stock_data['Close'], label=f'{stock_symbol} Closing Price') #gets the stuff its going to plot
plt.title(f'{stock_symbol} Stock Price During Covid') #makes tge title
plt.xlabel('Date') #x labe
plt.ylabel('Closing Price') # y label
plt.legend()
plt.show() #puts the information and outputs it
```

```python
# I put it all togetehr into one graph to see when covid havent
stock_symbols = ['AAL', 'PFE', 'HLT', 'F', "STLD", 'UAL', 'MAR'] #American Airliens, Phizer, Hilton Hotels, Ford, Steel Dynamics, United airlines, MARRIOT
start_date = '2020-01-01'
end_date = '2022-01-01'
stock_data = yf.download(stock_symbols, start=start_date, end=end_date)
plt.figure(figsize=(15,13))
sns.set(style="whitegrid")
for stock_symbol in stock_symbols:
    plt.plot(stock_data['Close'][stock_symbol], label=f'{stock_symbol} Closing Price')
plt.title('Stock Prices During Covid')
plt.xlabel('Date')
plt.yticks(rotation=50)
plt.gca().yaxis.set_major_locator(plt.MaxNLocator(nbins=38)) #makes the price on the y axis more specific makes analysis easier
plt.ylabel('Closing Price')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5)) # moves the lines color and name to the left of the graph
plt.show()

#They all go down at the same time, but go up at a different rate after first couple months of covid
```

```python
#finds the percentage of loss in prime time of covid
stock = ['AAL', 'PFE', 'HLT', 'F', "STLD", 'UAL', 'MAR']
for stock_symbol in stock:
    start_date = '2020-02-02'
    end_date = '2020-04-10'  # Adjust the end date as needed
    stock_data = yf.download(stock_symbols, start=start_date, end=end_date)
    closing_prices = stock_data['Close'][stock_symbol]
    percentage_loss = ((closing_prices.iloc[-1] - closing_prices.iloc[0]) / closing_prices.iloc[0]) * 100
    print(stock_symbol)
    print(percentage_loss)
```

```python
stock_symbols = ['AAL', 'UAL', 'HLT', 'MAR']
start_date = '2020-01-01'
end_date = '2021-01-01'
stock_data = yf.download(stock_symbols, start=start_date, end=end_date)
plt.figure(figsize=(12, 8)) #size of plot
sns.set(style="whitegrid")
for stock_symbol in stock_symbols:
    plt.hist(stock_data['Close'][stock_symbol], bins=30, alpha=0.5, label=stock_symbol) # this loop goes through each of the stocks
plt.title('Closing Prices Distribution for Selected Stocks')
plt.xlabel('Closing Price (USD)')
plt.ylabel('Frequency')
plt.legend()
```

```python
stock_symbols = ['AAL', 'UAL']
start_date = '2020-01-01'
end_date = '2021-01-01'
stock_data = yf.download(stock_symbols, start=start_date, end=end_date)
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
plt.scatter(stock_data.index, stock_data['Close']['AAL'], label='AAL Closing Prices', alpha=0.7, c='blue')
plt.scatter(stock_data.index, stock_data['Close']['UAL'], label='UAL Closing Prices', alpha=0.7, c='green')
plt.title('Scatter Plot of Closing Prices for AAL and UAL Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.show()
```
